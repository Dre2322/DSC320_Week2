```python
# 7
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Open the file car_data.csv and read it into a Pandas DataFrame.
cars = pd.read_csv('car_data.csv')

# pull hwy mpg from car_data.csv 
x = pd.DataFrame(cars, columns=['hwy_mpg'])
# pull weight from car_data.csv
y = pd.DataFrame(cars, columns=['weight'])

# create scatterplot
plt.scatter(x, y)

# Graph a line
def f(x):
    return -x * 95 + 5500

plt.plot(x, f(x), color='green')

# Graph a line using scilearns linear regression model
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x) 
plt.plot(x, y_pred, color='red')

# RMSE formula
def rmse(y, y_pred):
    return np.sqrt(np.mean((np.array(y) - np.array(y_pred))**2))

# Find RMSE for predicted line
rmse_error = rmse(f(x), y_pred)
# Finf RMSE for best-fit line
rmse_bestfit = rmse(y, y_pred)

# print RMSE results
print('RMSE for predicted line: ', rmse_error)
print(f'Best-fit RMSE: {rmse_bestfit}') 

# Best-fit line 
#line = np.polyfit(x.values.flatten(), y, 1)
#y_fit = np.polyval(line, x)
#plt.plot(x, y_fit, color='green')

# add labels 
plt.xlabel('hwy mpg')
plt.ylabel('weight')

plt.show()
```

```python
# 8
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Open the file car_data.csv and read it into a Pandas DataFrame.
cars = pd.read_csv('car_data.csv')

# pull hwy mpg from car_data.csv 
x = pd.DataFrame(cars, columns=['hwy_mpg'])
# pull weight from car_data.csv
y = pd.DataFrame(cars, columns=['weight'])

# create scatterplot
plt.scatter(x, y)

# Graph a line
def f(x):
    return -x * 95 + 5500

plt.plot(x, f(x), color='green')

# Graph a line using scilearns linear regression model
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x) 
plt.plot(x, y_pred, color='red')

# RMSE formula
def rmse(y, y_pred):
    return np.sqrt(np.mean((np.array(y) - np.array(y_pred))**2))

# Find RMSE for predicted line
rmse_error = rmse(f(x), y_pred)
# Finf RMSE for best-fit line
rmse_bestfit = rmse(y, y_pred)

# print RMSE results
print('RMSE for predicted line: ', rmse_error)
print(f'Best-fit RMSE: {rmse_bestfit}') 

# Fit a linear regression model 
pt = np.polyfit(y.values.flatten(), x.values.flatten(), 1)
# Predict the mpg for a vehicle weighing 3200lbs
pred_mpg = pt[0] * 3200 + pt[1]

print()
print(f'Predicted mpg for 3200lbs vehicle : {pred_mpg}')

# add labels 
plt.xlabel('hwy mpg')
plt.ylabel('weight')

plt.show()
```
